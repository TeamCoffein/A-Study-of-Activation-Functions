```
import numpy as np
import matplotlib.pylab as plt
```

# Heaviside Function

$$
\Theta(x) =  \begin{cases}
    0: & x < 0\\
    1: & x \geq 0
\end{cases}
$$

```
def heaviside(x):
    y = []
    for i in x:
        y.append(0 if i < 0 else 1)
    return y
```

```
x = np.linspace(-1, 1, 200)
y = heaviside(x)

# plot setup
plt.figure(figsize=(10,6))
ax1 = plt.subplot(111)
ax1.title.set_text('Heaviside Function')
ax1.grid(True, axis='both', linestyle='--')

ax1.plot([-1, 0], [0, 0], color='C0', label='Heaviside')
ax1.plot([0, 1], [1, 1], color='C0', label='Heaviside')
ax1.plot([0, 0], [0, 1], '--', color='C0', label='Heaviside')

ax1.plot(0, 1, 'o', c='C0')


ax1.set_ylim([-0.1, 1.1])
ax1.set_xlabel("x")
ax1.set_ylabel("y")

plt.show
```

# Sigmoid Activation Function

$$
S(x) = \frac{1}{1+e^{-x}} = \frac{e^x}{1+e^x}
$$

```
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```
x = np.linspace(-6, 6, 1200)
y = sigmoid(x)

# plot setup
plt.figure(figsize=(10,6))
ax1 = plt.subplot(111)
ax1.title.set_text('Sigmoid Function')
ax1.grid(True, axis='both', linestyle='--')

ax1.plot(x, y, label='Sigmoid')

ax1.set_ylim([-0.1, 1.1])
ax1.set_xlabel("x")
ax1.set_ylabel("y")

plt.show
```

# Rectivied Linerar Unit (Recitfier)

$$
f(x) = max(0, x)
$$

```
def relu(x):
    return np.maximum(0, x)
```

```
x = np.linspace(-1.3, 1.3, 200)
y = relu(x)

# plot setup
plt.figure(figsize=(10,6))
ax1 = plt.subplot(111)
ax1.title.set_text('Rectified Linear Unit')
ax1.grid(True, axis='both', linestyle='--')

ax1.plot(x, y, label='ReLU')

ax1.set_ylim([-0.1, 1.1])
ax1.set_xlabel("x")
ax1.set_ylabel("y")

plt.show
```

# Leaky Rectived Linerar Unit (Leaky ReLU)

$$
f(x) =  \begin{cases}
    0.01x: & x < 0\\
    x: & x \geq 0
\end{cases}
$$

```
def leaky_relu(x):
    y = []
    for i in x:
        y.append(0.01*i if i < 0 else i)
    return y
```

```
x = np.linspace(-1.3, 1.3, 200)
y = leaky_relu(x)

# plot setup
plt.figure(figsize=(10,6))
ax1 = plt.subplot(111)
ax1.title.set_text('Leaky Rectified Linear Unit')
ax1.grid(True, axis='both', linestyle='--')

ax1.plot(x, y, label='ReLU')

ax1.set_ylim([-0.1, 1.1])
ax1.set_xlabel("x")
ax1.set_ylabel("y")

plt.show
```

# Hyperbolic Tangent

$$
f(x) = tanh(x)
$$

```
def tanh(x):
    return np.tanh(x)
```

```
x = np.linspace(-3, 3, 200)
y = tanh(x)

# plot setup
plt.figure(figsize=(10,6))
ax1 = plt.subplot(111)
ax1.title.set_text('Hyperbolic Tangent')
ax1.grid(True, axis='both', linestyle='--')

ax1.plot(x, y, label='tanh')

ax1.set_ylim([-1.1, 1.1])
ax1.set_xlabel("x")
ax1.set_ylabel("y")

plt.show
```

# Squashing Function

$$
S_{\alpha, \lambda}^{(\beta)} = 
\frac{1}{\lambda\beta} \ln \frac{1+e^{\beta(x-(\alpha-\lambda/2))}}{1+e^{\beta(x-(\alpha+\lambda/2))}}
$$

```
def squashing(x, beta):
    alpha = 0.5
    lam = 1
    return (1/(lam*beta)) * np.log((1+np.exp(beta*(x-(alpha-lam/2)))) / (1+np.exp(beta*(x-(alpha+lam/2)))))
```

```
x = np.linspace(-3, 3, 200)
y1 = squashing(x, 0.5)
y2 = squashing(x, 1)
y3 = squashing(x, 2)
y4 = squashing(x, 5)
y5 = squashing(x, 100)


# plot setup
plt.figure(figsize=(10,6))
ax1 = plt.subplot(111)
ax1.title.set_text('Squashing Function')
ax1.grid(True, axis='both', linestyle='--')

ax1.plot(x, y1, label=' = 0.5')
ax1.plot(x, y2, label=' = 1')
ax1.plot(x, y3, label=' = 2')
ax1.plot(x, y4, label=' = 5')
ax1.plot(x, y5, label=' = 50')

ax1.set_ylim([-0.1, 1.1])
ax1.set_xlabel("x")
ax1.set_ylabel("y")
#ax1.legend()

plt.show
```
